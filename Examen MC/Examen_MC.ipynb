Parte 1:Carga y Limpieza (Pandas y NumPy)


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/diamonds.csv"
df = pd.read_csv(url)

print("Últimas 5 filas:")
print(df.tail(5))
print("\nResumen estadístico de columnas numéricas:")
print(df.describe())

df = df[(df['x'] != 0) & (df['y'] != 0) & (df['z'] != 0)]

df['volumen_aprox'] = df['x'] * df['y'] * df['z']

print("\nPrimeras filas con la nueva columna:")
print(df[['x', 'y', 'z', 'volumen_aprox']].head())

```

Parte 2: Visualización de Datos (Matplotlib)

```python
plt.figure(figsize=(8, 6))
plt.boxplot(df['price'])
plt.title('Distribución de Precios (Identificación de Outliers)')
plt.ylabel('Precio')
plt.show()

plt.figure(figsize=(10, 6))
plt.scatter(df['volumen_aprox'], df['price'], alpha=0.5)
plt.title('Relación entre Volumen Aproximado y Precio')
plt.xlabel('Volumen Aproximado')
plt.ylabel('Precio')
plt.show()

conteo_cortes = df['cut'].value_counts()

plt.figure(figsize=(10, 6))
conteo_cortes.plot(kind='bar')
plt.title('Cantidad de piezas por tipo de corte')
plt.xlabel('Tipo de Corte')
plt.ylabel('Cantidad de Piezas')
plt.show()
```
